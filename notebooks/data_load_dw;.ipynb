```python
dim_unidade = pd.read_csv(path_unidade, sep='\t')
dim_unidade.head()
```

```python
engine = connect_to_database()
table_name = "fato_lancamento"
fk_key = "pk_lancamento"
```

```python

def data_load(fk_key, table_name):
    with engine.connect() as connection:
        for _, row in dim_unidade.iterrows():
            select_verifica = f"SELECT {fk_key} FROM {table_name} WHERE {fk_key} = :fk_key"
            result = connection.execute(text(select_verifica), {'fk_key': row[fk_key]}).fetchall()

            if result: 
                print(f'O resultado já existe {row[fk_key]}')
            else: 
                print(f'Pode cadastrar no banco {row[fk_key]}')
                df = pd.DataFrame([row])
                df.to_sql(table_name, engine, if_exists='append', index=False)
```

```python
data_load(fk_key, table_name)
```
